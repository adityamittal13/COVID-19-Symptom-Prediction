```python
import pandas as pd

age_sympt = age_sympt.reset_index(drop = True)
age_sympt
```

```python
dic = {'0-17': {'symptom_names':[], 'counter_values':[]}, 
       '18-49':{'symptom_names':[], 'counter_values':[]}, 
       '50-64':{'symptom_names':[], 'counter_values':[]},
       '65+':{'symptom_names':[], 'counter_values':[]}}
for index in age_sympt.index:
    age = age_sympt['age'][index]
    age_symptoms = age_sympt['symptom'][index].split(', ')
    if (age >= 0 and age <= 17):
        for ind in age_symptoms:
            if (ind in dic['0-17']['symptom_names']):
                dic['0-17']['counter_values'][dic['0-17']['symptom_names'].index(ind)] += 1
            else:
                dic['0-17']['counter_values'].append(1)
                dic['0-17']['symptom_names'].append(ind)
    if (age >= 18 and age <= 49):
        for ind in age_symptoms:
            if (ind in dic['18-49']['symptom_names']):
                dic['18-49']['counter_values'][dic['18-49']['symptom_names'].index(ind)] += 1
            else:
                dic['18-49']['counter_values'].append(1)
                dic['18-49']['symptom_names'].append(ind)
    if (age >= 50 and age <= 64):
        for ind in age_symptoms:
            if (ind in dic['50-64']['symptom_names']):
                dic['50-64']['counter_values'][dic['50-64']['symptom_names'].index(ind)] += 1
            else:
                dic['50-64']['counter_values'].append(1)
                dic['50-64']['symptom_names'].append(ind)
    else:
        for ind in age_symptoms:
            if (ind in dic['65+']['symptom_names']):
                dic['65+']['counter_values'][dic['65+']['symptom_names'].index(ind)] += 1
            else:
                dic['65+']['counter_values'].append(1)
                dic['65+']['symptom_names'].append(ind)
    
print(dic['0-17'])
for ind in list(dic.keys()):
    dic[ind] = pd.DataFrame(data = dic[ind])

dic['0-17'] = dic['0-17'].sort_values('counter_values', ascending=False)
dic['18-49'] = dic['18-49'].sort_values('counter_values', ascending=False)
dic['50-64'] = dic['50-64'].sort_values('counter_values', ascending=False)
dic['65+'] = dic['65+'].sort_values('counter_values', ascending=False)
dic['18-49']
```

```python
dic['0-17'].to_csv('0-17.csv')
dic['18-49'].to_csv('18-49.csv')
dic['50-64'].to_csv('50-64.csv')
dic['65+'].to_csv('65+.csv')
```
