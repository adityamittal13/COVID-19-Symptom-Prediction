```python
import pandas as pd

age_sympt = age_sympt.reset_index(drop = True)
age_sympt
```

```python
dic = {'0-17': {'symptom_names':[], 'counter_values':[]}, 
       '18-49':{'symptom_names':[], 'counter_values':[]}, 
       '50-64':{'symptom_names':[], 'counter_values':[]},
       '65+':{'symptom_names':[], 'counter_values':[]}}
for index in age_sympt.index:
    age = age_sympt['age'][index]
    age_symptoms = age_sympt['symptom'][index].split(', ')
    if (age >= 0 and age <= 17):
        for ind in age_symptoms:
            if (ind in dic['0-17']['symptom_names']):
                dic['0-17']['counter_values'][dic['0-17']['symptom_names'].index(ind)] += 1
            else:
                dic['0-17']['counter_values'].append(1)
                dic['0-17']['symptom_names'].append(ind)
    if (age >= 18 and age <= 49):
        for ind in age_symptoms:
            if (ind in dic['18-49']['symptom_names']):
                dic['18-49']['counter_values'][dic['18-49']['symptom_names'].index(ind)] += 1
            else:
                dic['18-49']['counter_values'].append(1)
                dic['18-49']['symptom_names'].append(ind)
    if (age >= 50 and age <= 64):
        for ind in age_symptoms:
            if (ind in dic['50-64']['symptom_names']):
                dic['50-64']['counter_values'][dic['50-64']['symptom_names'].index(ind)] += 1
            else:
                dic['50-64']['counter_values'].append(1)
                dic['50-64']['symptom_names'].append(ind)
    else:
        for ind in age_symptoms:
            if (ind in dic['65+']['symptom_names']):
                dic['65+']['counter_values'][dic['65+']['symptom_names'].index(ind)] += 1
            else:
                dic['65+']['counter_values'].append(1)
                dic['65+']['symptom_names'].append(ind)
    
print(dic['0-17'])
for ind in list(dic.keys()):
    dic[ind] = pd.DataFrame(data = dic[ind])

dic['0-17'] = dic['0-17'].sort_values('counter_values', ascending=False)
dic['18-49'] = dic['18-49'].sort_values('counter_values', ascending=False)
dic['50-64'] = dic['50-64'].sort_values('counter_values', ascending=False)
dic['65+'] = dic['65+'].sort_values('counter_values', ascending=False)
dic['18-49']
```

```python
dic['0-17'].to_csv('0-17.csv')
dic['18-49'].to_csv('18-49.csv')
dic['50-64'].to_csv('50-64.csv')
dic['65+'].to_csv('65+.csv')
```

```python
from datetime import date

df_symptoms = df.dropna(subset = ['age', 'hosp_visit_date', 'exposure_start'])

#clustering algorithm based on how long it takes to go to the hospital after first symptom onset versus the age

df_symptoms = df_symptoms[['age', 'hosp_visit_date', 'exposure_start']]
df_symptoms = df_symptoms.reset_index(drop = True)
df_symptoms
#year month day
dictionary = {'age':[], 'numdays':[]}
for index in df_symptoms.index:
    hosp_date = str(df_symptoms['hosp_visit_date'][index]).split("/")
    exposure_date = str(df_symptoms['exposure_start'][index]).split("/")
    if (hosp_date[2] == "20"):
        hosp_date[2] = "2020"
    if (exposure_date[2] == "20"):
        exposure_date[2] = "2020"
    f_date = date(int(hosp_date[2]), int(hosp_date[0]), int(hosp_date[1]))
    l_date = date(int(exposure_date[2]), int(exposure_date[0]) , int(exposure_date[1]))
    delta = l_date - f_date
    dictionary['numdays'].append(abs(delta.days))
    dictionary['age'].append(df_symptoms['age'][index])

df_symptoms_new = pd.DataFrame(data = dictionary)
df_symptoms_new
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_symptoms_new['age'], df_symptoms_new['numdays'])
```

```python
#incubation period - time till you see the first symtoms
#graph above - gives you teh time from first symptoms to hospitalization
#Elbow Curve
from sklearn.cluster import KMeans

clusters = range(1, 20)
kmeans_elbow = [KMeans(n_clusters=i) for i in clusters]
score = [kmeans_elbow[i].fit(df_symptoms_new).score(df_symptoms_new) for i in range(len(kmeans_elbow))]
plt.plot(clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#Adding to Dataframe
df_symptoms_new["cluster_label"] = kmeans_elbow[5].fit(df_symptoms_new).labels_

#Applying K-Means Clustering (4 clusters)
kmeans = KMeans(n_clusters=6).fit(df_symptoms_new)
centroids = kmeans.cluster_centers_

plt.scatter(df_symptoms_new['age'], df_symptoms_new['numdays'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```
