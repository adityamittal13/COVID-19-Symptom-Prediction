```python
esymptoms = pd.read_csv ('COVID19_line_list_data.csv')
esymptoms= esymptoms.dropna(subset = ['symptom'])
esymptoms= esymptoms.dropna(subset = ['age'])
esymptoms
```

```python
ages=esymptoms[['age']]
print(ages)
```

```python
symptoms= esymptoms[['symptom']]
print(symptoms)
```

```python
age_sympt=pd.concat([ages, symptoms], axis=1)
age_sympt
```

```python
age_sympt.to_csv('age_sympt.csv',index=False)
```
